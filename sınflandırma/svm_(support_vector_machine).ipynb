# **2. veri önişleme**

**2.1 verilerin yüklenmesi**

```python
veriler = pd.read_csv("veriler.csv")
```

**test**

```python
veriler
```

```python
x = veriler.iloc[:,1:4].values # bağımsız [5:,1:4] en düzgün çalışan sınıflandırma temizlenmiş veri
y = veriler.iloc[:,4:].values
```

**verilerin eğitim ve test için bölünmesi**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test ,y_train , y_test = train_test_split(x , y , test_size= 0.33 , random_state = 0)
```

**verilerin ölçeklenmesi**

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

# ***buradan itibaren sınıflandırma algoritmaları başlar***

***1. logistic regression***

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression(random_state=0)
```

```python
logr.fit(X_train, y_train) # eğitim
```

```python
y_pred = logr.predict(X_test) #tahmin
y_pred
```

***karmaşık matrisi***

```python
cm = confusion_matrix(y_test , y_pred)
cm
```
