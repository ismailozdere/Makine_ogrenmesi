```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

# ***veri önişleme***

***veri yükleme***

```python
veri = pd.read_csv("odev_veri_kumesi.csv")
veri
```

# ***outlook encoder : kategorik -> numerik***

```python
outlook = veri.iloc[:,:1].values
outlook
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
outlook[:,0] =  le.fit_transform(veri.iloc[:,0])
outlook
```

```python
ohe = preprocessing.OneHotEncoder()
outlook = ohe.fit_transform(outlook).toarray()
outlook
```

***windy encoder***

```python
windy = veri.iloc[:,3:4].values
windy
```

```python
le = preprocessing.LabelEncoder()
```

```python
#windy[:,-1] = le.fit_transform(veri.iloc[:,-1])
windy
```

```python
windy = ohe.fit_transform(windy).toarray()
windy
```

***play encoder***

```python
play = veri.iloc[:,-1:].values
play
```

```python
play[:,-1] = le.fit_transform(veri.iloc[:,-1])
play
```

```python
play = ohe.fit_transform(play).toarray()
play
```

***sıcaklık ve nem encoder***

```python
temp_hum = veri.iloc[:,1:3].values
temp_hum
```

# ***numpy dizilerini dataframe dönüşümü***

```python
sonuc_o = pd.DataFrame(data= outlook ,index= range(14) ,columns=["sunny","overcast","rainy"])
sonuc_o
```

```python
sonuc_w = pd.DataFrame(data=windy[:,:1] , index= range(14), columns=["windy"])
sonuc_w
```

```python
sonuc_p = pd.DataFrame(data= play[:,:1] , index= range(14), columns=["no"])
sonuc_p
```

```python
sonuc_th = pd.DataFrame(data=temp_hum , index= range(14), columns=["temperature","humidity"])
sonuc_th
```

# ***dataframe birleştirme işlemi***

```python
s = pd.concat([sonuc_o , sonuc_th], axis=1)
s
```

```python
s2 = pd.concat([s , sonuc_w], axis=1)
s2
```

```python
s3 = pd.concat([s2 , sonuc_p], axis= 1)
s3
```

# ***verilerin eğitim ve test için bölünmesi***

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(s2 ,sonuc_p , test_size = 0.33 , random_state = 0 )
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train , y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```
